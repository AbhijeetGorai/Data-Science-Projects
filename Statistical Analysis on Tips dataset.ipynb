```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
tips_data = pd.read_csv("/content/tips.csv")
```

```python
tips_data.head()
```

**Checking the volume of customers in different days**

```python
df = pd.crosstab(tips_data['day'],tips_data['time'])
```

Checking the data

```python
df.head()
```

```python
Observed_values = df.values
```

Calculating Expected values based on Chi Square Formula

```python
val = stats.chi2_contingency(df)
val
```

```python
Expected_values = val[3]
```

Calculating degree of freedom

```python
alpha_value = 0.05
degree_of_freedom = (len(df.index)-1)*(len(df.columns)-1)
print("Degree of Freedom :",degree_of_freedom)
print("Alpha Value :",alpha_value)
```

Calculating Chi Square value

```python
chi_square = sum([(o-e)**2./e for o,e in zip(Observed_values,Expected_values)])
chi_square_statistic = chi_square[0] + chi_square[1]
print("Chi Square Statistic :",chi_square_statistic)
```

Calculating P value 

```python
p_value = 1-stats.chi2.cdf(x=chi_square_statistic,df=degree_of_freedom)
print("P value :",p_value)
```

```python
if p_value <= alpha_value:
  print("The volume of customers in all the days is different")
else:
  print("The voume of customers in all the days is same")
```

**Checking the tip paid by Male and Female**

```python
person = tips_data['sex']
tips = tips_data['tip']
```

```python
male_tip = person == 'Male'
female_tip = person == 'Female'
```

```python
male_tip_score = tips[male_tip].dropna()
female_tip_score = tips[female_tip].dropna()
```

Calculating T test value and p value

```python
T_val,p_val=stats.ttest_ind(male_tip_score,female_tip_score)
print("T test value is :",T_val)
print("P value is :",p_val)
```

```python
if p_val <= 0.05:
  print("Average tip paid by male and female is different")
else:
  print("Average tip paid by male and female is same")
```

**Checking relation between total bill and size of the group**

```python
corr_value,p = stats.pearsonr(tips_data['total_bill'],tips_data['size'])
print("Correlation value is :",corr_value)
print("P value is :",p)
```

```python
if p <= 0.05:
  print("There is linear relation between the Total Bill and size of thr group")
else:
  print("There is no linear relation between the Total bill and size of the group")
```
