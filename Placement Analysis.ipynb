```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
placement_data = pd.read_csv("/content/placement.csv")
```

Count of students in Different Streams

```python
plt.figure(figsize=(6,4))
sns.countplot("Degree stream",data=placement_data)
plt.xlabel("Stream")
plt.ylabel("Count of Students")
plt.show()
```

Checking the distribution in percentage of Placed and Not Placed students

```python
s = placement_data['status']
deg = placement_data['Mba %']
```

```python
placed_p = s == 'Placed'
notplaced_p = s == 'Not Placed'
```

```python
placed_score = deg[placed_p].dropna()
notplaced_score = deg[notplaced_p].dropna()
```

Calculating P value and T test score

```python
T_val,p_val=stats.ttest_ind(placed_score,notplaced_score)
print("T test value is :",T_val)
print("P value is :",p_val)
```

Checking distribution of the data

```python
if p_val <= 0.05:
  print("Degree percentage of placed and non placed students are different")
else:
  print("Degree percentage of placed and non placed students are same")
```
